# TAREA 8 MÍNIMOS CUADRADOS

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

#Esta función realiza el método de mínimos cuadrados, grafica y calcula el error cuadrático

def minimoscuadrados(x,y,g):    
    coef = np.polyfit(x, y, g)  
    print('Coeficientes del Polinomio')
    print(coef)
    
    plt.figure(figsize=(10, 6))
    plt.scatter(x, y, color="red", label="Datos originales")

    y_ajustado = np.polyval(coef, x)

    plt.plot(x, y_ajustado, color="blue", label=f"Ajuste")
    plt.legend()
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("Ajuste por mínimos cuadrados")
    plt.grid()
    plt.show()


    # Calcular el Error Cuadrático Medio
    ECM = mean_squared_error(y, y_ajustado)
    print(f"Error Cuadrático Medio: {ECM}")
```

### EJERCICIO 1

```python
xi = np.array([4.0, 4.2, 4.5, 4.7, 5.1, 5.5, 5.9, 6.3, 6.8, 7.1])
yi =np.array([102.56, 130.11, 113.18, 142.05, 167.53, 195.14, 224.87, 256.73, 299.50, 326.72])
```

a. Con grado 1:

```python
minimoscuadrados(xi,yi,1)
```

b. Con grado 2

```python
minimoscuadrados(xi,yi,2)
```

c. Con grado 3

```python
minimoscuadrados(xi,yi,3)
```

d. Forma $be^{ax}$

```python
z = np.log(yi)

coef = np.polyfit(xi, z, 1)
a = coef[0]  
ln_b = coef[1]  
b = np.exp(ln_b)  
print('Coeficientes del Polinomio')
print(coef)
# Construcción del modelo ajustado
y_ajustado = b * np.exp(a * xi)

mse = mean_squared_error(yi, y_ajustado)
rmse = np.sqrt(mse)

plt.figure(figsize=(10, 6))
plt.scatter(xi, yi, color="red", label="Datos originales")
plt.plot(xi, y_ajustado, color="blue", label=f"Ajuste")
plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Ajuste por mínimos cuadrados")
plt.grid()
plt.show()

ECM = mean_squared_error(yi, y_ajustado)
print(f"Error Cuadrático Medio: {ECM}")
```

e. Forma $bx^{a}$

```python
log_x = np.log(xi)
log_y = np.log(yi)

coef = np.polyfit(log_x, log_y, 1)  
a = coef[0]  # Pendiente
ln_b = coef[1]  # Intersección
b = np.exp(ln_b)
print('Coeficientes del Polinomio')
print(coef)

y_ajustado = b * xi**a

mse = mean_squared_error(yi, y_ajustado)
rmse = np.sqrt(mse)

plt.figure(figsize=(10, 6))
plt.scatter(xi, yi, color="red", label="Datos originales")
plt.plot(xi, y_ajustado, color="blue", label=f"Ajuste")
plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Ajuste por mínimos cuadrados")
plt.grid()
plt.show()

ECM = mean_squared_error(yi, y_ajustado)
print(f"Error Cuadrático Medio: {ECM}")
```
